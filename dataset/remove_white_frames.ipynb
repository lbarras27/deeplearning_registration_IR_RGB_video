```python
import util
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = "DATA3/videos6/"

# load the images
imgs_rgb, imgs_ir = util.loadDataset(path, "RGB_final", "IR_final")
imgs_ir.shape
```

```python
imgs_rgb_mask, imgs_ir_mask = util.loadDataset(path, "RGB_mask", "IR_mask", color=False)
imgs_rgb_mask.shape
```

```python
rgb_final = []
ir_final = []
mask_rgb_final = []
mask_ir_final = []

for i in range(0, imgs_rgb_mask.shape[0]):
    mask = imgs_rgb_mask[i]
    if (len(mask[mask == 255])/(320*240)) == 1.0:
        print("remove frame" + str(i))
    else:
        rgb_final.append(imgs_rgb[i])
        ir_final.append(imgs_ir[i])
        mask_rgb_final.append(imgs_rgb_mask[i])
        mask_ir_final.append(imgs_ir_mask[i])

rgb_final = np.array(rgb_final)
ir_final = np.array(ir_final)
mask_rgb_final = np.array(mask_rgb_final)
mask_ir_final = np.array(mask_ir_final)
```
