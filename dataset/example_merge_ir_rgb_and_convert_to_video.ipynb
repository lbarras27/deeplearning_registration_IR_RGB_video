```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import re
import sys
sys.path.append('../')

import util

%matplotlib inline
```

```python
path = "../imgs/"

# load the rgb and ir frames
imgs_rgb, imgs_ir = util.loadDataset(path, "rgb", "ir")
```

```python
no = 5
# superpose the 'no' rgb and ir frame
res = util.superposeTwoImages(imgs_rgb[no], imgs_ir[no])
```

```python
# print the results
plt.figure(figsize=(15, 15))
plt.subplot(131)
plt.title("RGB image")
plt.imshow(cv2.cvtColor(imgs_rgb[no], cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title("IR image")
plt.imshow(cv2.cvtColor(imgs_ir[no], cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.title("Superposition RGB and IR")
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
# superposes all the rgb and ir frames
imgs_to_video = util.superposeTwoImages(imgs_rgb, imgs_ir)
```
